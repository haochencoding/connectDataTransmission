# Functions for retrieval connect data

```python
def retrieve_data(data_class, connect_session_token, skip=0):
    # Constants
    CONNECT_URL = 'https://connect-project.io'
    PARSE_APPLICATION = 'connect'

    #Parameters for requests
    url = f'{CONNECT_URL}/parse/classes/{data_class}'
    headers = {
        'x-parse-application-id': PARSE_APPLICATION,
        'x-parse-session-token': connect_session_token,
    }

    params = {
        'limit': '100',
        'skip': str(skip),
    }

    # Retrieve data
    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()  # Raises an HTTPError for bad responses
    data = response.json()['results']

    return data
```

```python
def retrieve_data_all(data_class, connect_session_token):
    print(f"{'-'*10}Retrieving {data_class} {'-'*10}")
    count=0
    data_all = []

    while True:
        data_temp = retrieve_data(data_class, connect_session_token, skip=count)
        if data_temp != []:
            data_all += data_temp
            count += len(data_temp)
            print(f"{len(data_temp)} records retrieved; total: {count} records")
        else:
            print(f'Retrieval complete! total: {count} records')
            break
    
    return data_all
```
